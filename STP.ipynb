```python
df_filtered = dtsnew[(dtsnew['STATUS'] == 'Operational') | (dtsnew['STATUS'] == 'Not Reported')]
```

```python
df_filtered.drop('WWTP_NAME', axis=1, inplace=True)
```

```python
df_filtered.info()
```

```python
df_filtered = df_filtered.drop(['COUNTRY', 'QUAL_LOC','STATUS','QUAL_POP','QUAL_LEVEL','QUAL_WASTE','QUAL_CAP'], axis=1)
```

```python
df_filtered.info()
```

```python
df_filtered.duplicated().sum()
```

```python
df_filtered.columns.tolist()
```

```python
df_filtered.info()
```

```python
df_filtered['DF'].fillna(df_filtered['DF'].median(), inplace=True)
```

```python
df_filtered['RIVER_DIS'].fillna(df_filtered['RIVER_DIS'].median(), inplace=True)
df_filtered['DESIGN_CAP'].fillna(df_filtered['DESIGN_CAP'].median(), inplace=True)
```

```python
df_filtered.info()
```

```python
columns_to_convert = ['LAT_WWTP','LON_WWTP','WASTE_DIS','DF','RIVER_DIS','DESIGN_CAP']
df_filtered[columns_to_convert] = df_filtered[columns_to_convert].astype(int)
```

```python
df_filtered = pd.get_dummies(df_filtered, columns=['LEVEL'], drop_first=True)
```

```python
df_filtered.info()
```

```python
df_filtered.head()
```

```python
df_filtered.shape
```

```python
df_filtered.duplicated().sum()
```

```python
df_filtered = df_filtered.drop_duplicates()
```

```python
df_filtered.shape
```

```python
df_filtered['EFFICIENCY'] = df_filtered['WASTE_DIS'] / df_filtered['DESIGN_CAP']
df_filtered['EFFICIENCY'].fillna(df_filtered['EFFICIENCY'].median(), inplace=True)
```

```python
df_filtered.info()
```

```python
df_filtered.replace([np.inf, -np.inf], np.nan, inplace=True)
df_filtered['EFFICIENCY'] = df_filtered['EFFICIENCY'].fillna(df_filtered['EFFICIENCY'].median())
df_filtered['EFFICIENCY'] = df_filtered['EFFICIENCY'].astype(int)
```

```python
x=df_filtered[['LAT_WWTP','LON_WWTP','POP_SERVED','DF','RIVER_DIS','DESIGN_CAP','LEVEL_Primary','LEVEL_Secondary']]
y=df_filtered[['EFFICIENCY']]
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor  
model = RandomForestRegressor(n_estimators=100, random_state=42)  
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score  
y_pred = model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RÂ² Score:", r2_score(y_test, y_pred))
```
